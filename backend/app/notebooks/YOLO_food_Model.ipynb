```python
import requests
import zipfile
import os

urls = {
    "coco": "https://zenodo.org/records/13633554/files/chownet_object_detection_coco_format.zip?download=1",
    "yolo": "https://zenodo.org/records/13633554/files/chownet_object_detection_yolo_format.zip?download=1"
}

download_dir = "downloads"
extract_dir = "datasets"

os.makedirs(download_dir, exist_ok=True)
os.makedirs(extract_dir, exist_ok=True)

def download_file(url, save_path):
    print(f"Downloading: {url}")
    response = requests.get(url, stream=True)
    response.raise_for_status()
    with open(save_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
    print(f"Saved to {save_path}")

def unzip_file(zip_path, extract_to):
    print(f"Extracting {zip_path} ...")
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"Extracted to {extract_to}")


for key, url in urls.items():
    zip_path = os.path.join(download_dir, f"{key}.zip")
    download_file(url, zip_path)
    target_path = os.path.join(extract_dir, key)
    unzip_file(zip_path, target_path)

print("All downloads and extractions completed.")
```

```python
import requests
import zipfile
import os

urls = {
    "coco": "https://zenodo.org/records/13633554/files/chownet_object_detection_coco_format.zip?download=1",
    "yolo": "https://zenodo.org/records/13633554/files/chownet_object_detection_yolo_format.zip?download=1"
}

download_dir = "downloads"
extract_dir = "datasets"

os.makedirs(download_dir, exist_ok=True)
os.makedirs(extract_dir, exist_ok=True)

def download_file(url, save_path):
    print(f"Downloading: {url}")
    response = requests.get(url, stream=True)
    response.raise_for_status()
    with open(save_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
    print(f"Saved to {save_path}")

def unzip_file(zip_path, extract_to):
    print(f"Extracting {zip_path} ...")
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"Extracted to {extract_to}")


for key, url in urls.items():
    zip_path = os.path.join(download_dir, f"{key}.zip")
    download_file(url, zip_path)
    target_path = os.path.join(extract_dir, key)
    unzip_file(zip_path, target_path)

print("All downloads and extractions completed.")
```

```python
import os

# Base paths
coco_base = "datasets/coco/chownet_object_detection_coco_format"
yolo_base = "datasets/yolo/chownet_object_detection_yolo_format"

# COCO contents
coco_images_dir = os.path.join(coco_base, "images")
coco_annotations_file = os.path.join(coco_base, "results.json")
coco_classes_file = os.path.join(coco_base, "classes.txt")

print("COCO images:", len(os.listdir(coco_images_dir)))
print("COCO annotation file exists:", os.path.exists(coco_annotations_file))
print("COCO classes file exists:", os.path.exists(coco_classes_file))

# YOLO contents
yolo_images_dir = os.path.join(yolo_base, "images")
yolo_labels_dir = os.path.join(yolo_base, "labels")
yolo_classes_file = os.path.join(yolo_base, "classes.txt")

print("YOLO images:", len(os.listdir(yolo_images_dir)))
print("YOLO labels:", len(os.listdir(yolo_labels_dir)))
print("YOLO classes file exists:", os.path.exists(yolo_classes_file))

# Check for missing labels or images in YOLO
image_files = set([f.split('.')[0] for f in os.listdir(yolo_images_dir)])
label_files = set([f.split('.')[0] for f in os.listdir(yolo_labels_dir)])
print("Images without labels:", image_files - label_files)
print("Labels without images:", label_files - image_files)
```

```python
import ultralytics
from ultralytics import YOLO

print("Yolo version: ", ultralytics.__version__)
```

```python
yolo_dataset_path = "datasets/yolo/chownet_object_detection_yolo_format"
images_path = os.path.join(yolo_dataset_path, "images")
labels_path = os.path.join(yolo_dataset_path, "labels")
classes_file = os.path.join(yolo_dataset_path, "classes.txt")

print("Number of images:", len(os.listdir(images_path)))
print("Number of labels:", len(os.listdir(labels_path)))
print("Classes file exists:", os.path.exists(classes_file))

with open(classes_file, "r") as f:
    classes = [line.strip() for line in f.readlines()]
print("Classes:", classes)
```

```python
import yaml
import os

yolo_yaml_path = "datasets/yolo/chownet_data.yaml"

# yolo_dataset_path = "datasets/yolo/chownet_object_detection_yolo_format"

data_yaml = {
    'train': os.path.abspath(os.path.join(yolo_dataset_path, "images")),
    'val': os.path.abspath(os.path.join(yolo_dataset_path, "images")),  # since dataset is small, use same for val
    'names': classes
}

with open(yolo_yaml_path, "w") as f:
    yaml.dump(data_yaml, f)

print(f"YOLO dataset YAML created at: {yolo_yaml_path}")
```

```python
from ultralytics import YOLO

data_yaml = "datasets/yolo/chownet_data.yaml"
model = YOLO("yolov8n.pt")

model.train(
    data=data_yaml, epochs=50, imgsz=640, batch=4, name="chownet_yolo", project="runs/train", exist_ok=True
)
```

```python
# Load best weights automatically saved by YOLO
best_model = YOLO("runs/train/chownet_yolo/weights/best.pt")

metrics = best_model.val()
print(metrics)
```
